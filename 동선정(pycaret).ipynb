```python
# read csv file
import pandas as pd
import numpy as np
data = pd.read_excel('전처리(22.04.03).xlsx')
```

```python
index1 = data[data['안심택배함수'] == 0].index
```

```python
data2 = data.drop(index1)
```

```python
from pycaret.regression import *
reg = setup(data2, target = '안심택배함수', train_size=0.8)
```

```python
##모델비교
best = compare_models(sort = 'RMSE')
```

```python
model = create_model(best, cross_validation = False)
```

```python
##하이퍼파라미터 튜닝
tuned_model = tune_model(model, optimize = 'RMSE', n_iter = 10)
```

```python
# 마지막 학습(Finalize)

final_model = finalize_model(tuned_model)


# 예측(Predict)

pred = predict_model(final_model, data = data)
```

```python
pred.to_excel("동선정(pycaret)(택배함 있는 동).xlsx")
```
